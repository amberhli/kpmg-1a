```python
temp_cols = [
    "BTU2(Gwell_L)_ReturnT (F)", "BTU2(Gwell_L)_SupplyT (F)",
    "BTU3(Gwell_R)_ReturnT (F)", "BTU3(Gwell_R)_SupplyT (F)",
    "BTU4(Gwell_Com)_ReturnT (F)", "BTU4(Gwell_Com)_SupplyT (F)"
]

energy_cols = [
    "BTU2(Gwell_L)_energy (BTUs)",
    "BTU3(Gwell_R)_energy (BTUs)",
    "BTU4(Gwell_Com)_energy (BTUs)"
]

flow_cols = [
    "BTU2(Gwell_L)_flow (GPM)",
    "BTU3(Gwell_R)_flow (GPM)",
    "BTU4(Gwell_Com)_flow (GPM)"
]
```

```python
for col in df_main.columns:
    if col in temp_cols:
        df_main.loc[((df_main[col] < 40) | (df_main[col] > 120)), col] = np.nan

        
```

```python
for col in df_main.columns:
    if col in energy_cols or col in temp_cols:
        df_main.loc[df_main[col] < 0, col] = np.nan
```

```python
df_main.isna().sum()
```

```python
df_main = df_main.sort_index()
```

```python
field = 'BTU4(Gwell_Com)_ReturnT (F)'
df_main[field] = df_main[field].interpolate(limit=2)
```

```python
df_main.isna().sum()
```

```python
scaler = StandardScaler()
df_scaled = df_main
df_scaled[temp_cols+energy_cols+flow_cols] = scaler.fit_transform(df_scaled[temp_cols + energy_cols + flow_cols])
```
